```python
con = duckdb.connect()
```

```python
create_table_query = """
CREATE TABLE products AS 
SELECT * FROM 
read_csv_auto('Data/product.csv');
"""
con.execute(create_table_query)
```

```python
select_query = """
            SELECT * FROM
            products;
            """
            
result_1 = con.execute(select_query).fetchall()

for row in result_1:
    print(row)
```

```python
create_table_query_2 = """
CREATE TABLE orders AS 
SELECT * FROM 
read_csv_auto('Data/orders.csv');
"""
con.execute(create_table_query_2)
```

```python
select_query_2 = """
            SELECT * FROM
            orders;
            """

result_2 = con.execute(select_query_2).fetchall()

for row in result_2:
    print(row)
    
    
```

```python
join_query = """
CREATE VIEW product_perf as 
SELECT p.Product_ID as Product_ID,
p.ProductName as Product_Name,
p.PRODUCTCATEGORY as product_category,
p.Price as Product_Price,
sum(o.Quantity) as Total_product_sold,
count(o.Order_ID) as Total_product_orders 
FROM 
products p
JOIN orders o 
ON 
o.ProductID = p.Product_Id
 
group by 1, 2, 3, 4
"""

result = con.execute(join_query).fetchall()


```
